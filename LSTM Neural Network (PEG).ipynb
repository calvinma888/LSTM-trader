```python
import numpy as np
import pandas as pd
```

```python
data= pd.read_csv("PEGtechnicals.csv")
```

```python
list(data.columns)
```

```python
import tensorflow as tf
from sklearn.model_selection import train_test_split
```

```python
Y = data["PEG_close"]
X = data.drop(['Unnamed: 0',"PEG_time",
 'PEG_Unnamed: 0',"PEG_close","PEG_momentum_kama"], axis=1)
```

```python
# Lagging the Y by 1 time step
# Predicting one step ahead
Y= Y[31:]
X= X[30:-1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.4, random_state=60)
```

```python
X_train
```

```python
X_train = X_train.dropna(axis =1)
```

```python
X_test = X_test.dropna(axis =1)
```

```python
X_test
```

```python
correlation_row = data.corr()[5:6]
```

```python
high_corr_list = []
for i in range(len(list(correlation_row.T["PEG_close"]))):
    if list(correlation_row.T["PEG_close"])[i] > 0.5 or list(correlation_row.T["PEG_close"])[i] < -0.5:
        high_corr_list.append(list(correlation_row)[i])
        print (f'{list(correlation_row)[i]} has a correlation with closing price of {list(correlation_row.T["PEG_close"])[i]}')
```

```python
high_corr_list
```

```python
# Drop Close 
high_corr_list.remove("PEG_close")

```

```python
my_list = ['PEG_trend_psar_down', 'PEG_trend_visual_ichimoku_a', 'PEG_momentum_kama', 'PEG_trend_psar_up']
for i in my_list:
    high_corr_list.remove(i)
```

```python
high_corr_list
```

```python
X_train3 = X_train[high_corr_list]
X_test3 = X_test[high_corr_list]
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Dense, LSTM
model = Sequential()
model.add(LSTM(100,input_shape=(X_train3.shape[1],1)))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
X_train3 = np.array(X_train3).reshape((X_train3.shape[0],X_train3.shape[1],1))
X_test3 = np.array(X_test3).reshape((X_test3.shape[0],X_test3.shape[1],1))
```

```python
history = model.fit(x=X_train3[:103606],y=Y_train[:103606],epochs=2,validation_data=(X_train3[103606:],Y_train[103606:]),shuffle=False)
```

```python
import pandas as pd
import yfinance as yf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams["figure.dpi"] = 300

axes=plt.axes()
axes.plot(pd.DataFrame(model.history.history)['loss'], label='Loss')
axes.plot(pd.DataFrame(model.history.history)['val_loss'], label='Validation Loss')
axes.legend(loc=0)
axes.set_title('PEG Model fitting performance')
```

```python
Y_predicted=model.predict(np.array(X_test3))
Y_true=np.array(Y_test).reshape(np.array(Y_test).shape[0],1)
```

```python
plt.rcParams["figure.dpi"] = 300
plt.plot(Y_true[::100], label='True Y')
plt.plot(Y_predicted[::100], label='Predicted Y')
plt.legend(loc=0)
plt.title('PEG LSTM Prediction vs Actual')
plt.ylabel("Price")
plt.show()
```

```python
# Calculating the profits from the model

actual_y = list(Y)
predicted_y1 = model.predict(np.array(X_train3))
predicted_y2 = model.predict(np.array(X_test3))
```

```python
predicted_y = [predicted_y1[i][0] for i in range(len(predicted_y1))] + [predicted_y2[i][0] for i in range(len(predicted_y2))]
```

```python
starting_value = 1000000
portfolio_value = [starting_value]
```

```python
stock_amount = portfolio_value[-1]/actual_y[0]
```

```python
holding_period = 1
```

```python
# 2.75 bsp transaction fees

def calculate_returns(holding_period, predicted_y, actual_y):
    starting_value = 1000000
    portfolio_value = [starting_value]
    
    for i in range(0,len(actual_y)-holding_period,holding_period):
        if predicted_y[i+holding_period] > actual_y[i]:
            
            #Buying if predicted price in the next time step is higher
            stock_amount = (portfolio_value[-1]*(1-0.000275))/actual_y[i]
            
            #Selling taking into account the transaction fees
            new_portfolio_value = (stock_amount*actual_y[i+holding_period])*(1-0.000275)
            portfolio_value.append(new_portfolio_value)
            
        elif predicted_y[i+holding_period] < actual_y[i]:
            
            #Selling first then buying back
            new_value_after_short = (- portfolio_value[-1]/actual_y[i] + portfolio_value[-1]/actual_y[i]*actual_y[i+holding_period])*(1-0.000275)**2
            portfolio_value.append(new_value_after_short)
        else:
            portfolio_value.append(portfolio_value[-1])
            
    return portfolio_value
```

```python
holding_period_1 = calculate_returns(1, predicted_y, actual_y)
```

```python
holding_period_5 = calculate_returns(5, predicted_y, actual_y)
```

```python
holding_period_10 = calculate_returns(10, predicted_y, actual_y)
```

```python
holding_period_1000 = calculate_returns(1000, predicted_y, actual_y)
```

```python
holding_period_5000 = calculate_returns(5000, predicted_y, actual_y)
```

```python
plt.rcParams["figure.dpi"] = 300
plt.plot(holding_period_1[:100], color='red', label= "holding for 1 minutes")
plt.plot(holding_period_5[:100], color='green', label= "holding for 5 minutes")
plt.plot(holding_period_10[:100], color='blue', label= "holding for 10 minutes")
plt.plot(holding_period_1000[:100], color='orange', label= "holding for 1000 minutes")
plt.plot(holding_period_5000[:100], color='pink', label= "holding for 5000 minutes")
plt.ylabel("Portfolio Value")

# Both Long and Short
plt.legend(loc=0)
plt.title('High Frequency Trading with LSTM Predictions: Different Holding Periods (PEG)\n')
plt.show()
```

```python
# 2.75 bsp transaction fees, only execute if predicted is 20% higher

def calculate_returns2(holding_period, predicted_y, actual_y):
    starting_value = 1000000
    portfolio_value = [starting_value]
    
    for i in range(0,len(actual_y)-holding_period,holding_period):
        if predicted_y[i+holding_period] > 1.20*actual_y[i]:
            
            #Buying if predicted price in the next time step is higher
            stock_amount = (portfolio_value[-1]*(1-0.000275))/actual_y[i]
            
            #Selling taking into account the transaction fees
            new_portfolio_value = (stock_amount*actual_y[i+holding_period])*(1-0.000275)
            portfolio_value.append(new_portfolio_value)
            
        elif 1.20*predicted_y[i+holding_period] < actual_y[i]:
            
            #Selling first then buying back
            new_value_after_short = (- portfolio_value[-1]/actual_y[i] + portfolio_value[-1]/actual_y[i]*actual_y[i+holding_period])*(1-0.000275)**2
            portfolio_value.append(new_value_after_short)
            
        else:
            portfolio_value.append(portfolio_value[-1])
    return portfolio_value
```

```python
holding_period_1_2 = calculate_returns2(1, predicted_y, actual_y)
holding_period_5_2 = calculate_returns2(5, predicted_y, actual_y)
holding_period_10_2 = calculate_returns2(10, predicted_y, actual_y)
holding_period_1000_2 = calculate_returns2(1000, predicted_y, actual_y)
holding_period_5000_2 = calculate_returns2(5000, predicted_y, actual_y)
```
